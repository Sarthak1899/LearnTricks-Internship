```python
cat_feats = ['InvoiceAmount_bin']
final_data = pd.get_dummies(data,columns=cat_feats,drop_first=True)
```

```python
features=['countryCode', 'customerID', 'InvoiceAmount',
       'Disputed', 'PaperlessBill','repeatCust','Late', 'DaysToSettle',
       'countlate']
```
