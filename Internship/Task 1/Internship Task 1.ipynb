<h2>Model Selection</h2>

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale
X= data.drop('Exited',axis =1)
X =scale(X)
Y = data['Exited']
X_train,x_test,Y_train,y_test = train_test_split(X,Y,test_size = 0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X_train,Y_train)
```

```python
linear_model.score(X_train,Y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_model = Lasso(alpha=0.01)
lasso_model.fit(X_train,Y_train)
```

```python
lasso_model.score(X_train,Y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(max_depth=15, min_samples_leaf=100)
dt_model.fit(X_train,Y_train)
```

```python
dt_model.score(X_train,Y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=200,max_depth=25, min_samples_leaf=60,n_jobs=4)
rf_model.fit(X_train,Y_train)
```

```python
rf_model.score(X_train,Y_train)
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 5,interaction_only=False, include_bias=True)
X_poly = poly_reg.fit_transform(X_train)
poly_reg.fit(X_train, Y_train)
lassolars_reg = linear_model.LassoLars(alpha=0.021,max_iter=151)
lassolars_reg.fit(X_poly, Y_train)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
def naive_bayes(x_train,y_train):
    classifier = GaussianNB()
    classifier.fit(x_train,y_train)
    
    return classifier
```

```python
def k_nearest_neighbors(x_train,y_train):
    classifier = KNeighborsClassifier(n_neighbors=10)
    classifier.fit(x_train,y_train)
    
    return classifier
```

```python
def svc(x_train,y_train):
    classifier = SVC(kernel = 'rbf', gamma = 'scale')
    classifier.fit(x_train,y_train)
    
    return classifier
```

```python
def decision_tree(x_train,y_train):
    classifier = DecisionTreeClassifier(max_depth=6)
    classifier.fit(x_train,y_train)
    
    return classifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
def build_and_train_classifier(x_train,y_train,classification_fn):
    model = classification_fn(x_train,y_train)
    y_pred = model.predict(x_test)
    
    train_score = model.score(x_train,y_train)
    test_score = accuracy = accuracy_score(y_test,y_pred)
    
    print("Training Score: ",train_score)
    print("Testing Score: ", test_score)
```

```python
build_and_train_classifier(X_train,Y_train,naive_bayes)
```

```python
build_and_train_classifier(X_train,Y_train,k_nearest_neighbors)
```

```python
build_and_train_classifier(X_train,Y_train,svc)
```

```python
build_and_train_classifier(X_train,Y_train,decision_tree)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
logreg = LogisticRegression()
logreg.fit(X_train,Y_train)
y_pred = logreg.predict(X_train)
print('Train accuracy score:',accuracy_score(Y_train,y_pred))
print('Test accuracy score:', accuracy_score(y_test,logreg.predict(x_test)))
```

<p><b>Since SVC gives us the highest accuracy we will go ahead with it</b></p>
